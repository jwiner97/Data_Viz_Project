```python
# Import Dependents
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Define Files
county_deaths = 'Resources/Multiple Cause of Death Opioits 2006-2017 (3).csv'
unemployment_rate06 = 'Resources/laucnty06.csv'
unemployment_rate07 = 'Resources/laucnty07.csv'
unemployment_rate08 = 'Resources/laucnty08.csv'
unemployment_rate09 = 'Resources/laucnty09.csv'
unemployment_rate10 = 'Resources/laucnty10.csv'
unemployment_rate11 = 'Resources/laucnty11.csv'
unemployment_rate12 = 'Resources/laucnty12.csv'
unemployment_rate13 = 'Resources/laucnty13.csv'
unemployment_rate14 = 'Resources/laucnty14.csv'
unemployment_rate15 = 'Resources/laucnty15.csv'
unemployment_rate16 = 'Resources/laucnty16.csv'
unemployment_rate17 = 'Resources/laucnty17.csv'
state_deaths17 = "Resources/KFF Opioid Deaths by State 2017.csv"
state_deaths_over_time = "Resources/Deaths 99-17.csv"

```

```python
unemployment_files = {
    2006: 'Resources/laucnty06.csv',
    2007: 'Resources/laucnty07.csv',
    2008: 'Resources/laucnty08.csv',
    2009: 'Resources/laucnty09.csv',
    2010: 'Resources/laucnty10.csv',
    2011: 'Resources/laucnty11.csv',
    2012: 'Resources/laucnty12.csv',
    2013: 'Resources/laucnty13.csv',
    2014: 'Resources/laucnty14.csv',
    2015: 'Resources/laucnty15.csv',
    2016: 'Resources/laucnty16.csv',
    2017: 'Resources/laucnty17.csv'
}
```

```python
dfs = []
for year, filename in unemployment_files.items():
    df = pd.read_csv(filename, skiprows=4)
    df['year'] = year
    dfs.append(df)
```

```python
full_data = pd.concat(dfs)
full_data.tail()
```

```python
full_data.set_index('County Name/State Abbreviation', inplace=True)
full_data.head()
```

```python
UE_df = full_data.loc[['Mohave County, AZ',
                                          'Cook County, IL',
                                          'Lake County, IN',
                                          'Iberia Parish, LA',
                                          'Genessee County, MI',
                                         'Saginaw County, MI',
                                         'Wayne County, MI',
                                         'Mahoning County, OH',
                                         'Richland County, OH',
                                         'Trumbull County, OH']]

UE_df = UE_df.drop(["Code", "Code.1", "Code.2", "Year", "Unnamed: 5"], axis=1)

UE_df.head()
```

```python
Primary_UE_df = UE_df.drop(["Force", "Employed", "Unemployed"], axis=1)
print(Primary_UE_df)
```

```python
peter_df = Primary_UE_df.reset_index()
peter_df.head()
```

```python
peter_df['(%)'] = peter_df['(%)'].astype('float64')
```

```python
p = peter_df.set_index('year')
```

```python
p.head(20)
```

```python
p.plot(kind='line')
```

```python
# pd.pivot_table(data, index= 'Year', values= "Happiness Score")
Unemployment = pd.pivot_table(peter_df, index='year', values='(%)', columns='County Name/State Abbreviation')
```

```python
Unemployment
```

```python
Unemployment.plot()
plt.legend(loc= 'center left', bbox_to_anchor=(1,.5))
plt.ylabel('Percent Unemployment')
plt.suptitle("Unemployment by County from 2006-2017", fontsize=14)
plt.savefig("Images/Unemployment.png")
plt.show()
```
